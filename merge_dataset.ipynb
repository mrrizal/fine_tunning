```python
import json
import os
import random
```

```python
filenames = [f for f in os.listdir('./synthetic_datasets') if f.endswith('.json')]
print(filenames)
```

```python
base_dir = './synthetic_datasets'
merged_data = []
for filename in filenames:
    with open(os.path.join(base_dir, filename), 'r') as f:
        data = json.load(f)
        merged_data.extend(data)
```

```python
print(f'Total merged records: {len(merged_data)}')
```

```python
random.shuffle(merged_data)
```

```python
evaluation_data = [merged_data.pop() for _ in range(30)]
```

```python
print(len(evaluation_data))
```
